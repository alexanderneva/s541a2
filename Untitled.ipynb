# Question 1: Quadratic Discriminant Analysis

We know from the textbook that 

$$
Pr(Y=k|X=x) = \frac{\pi_kf_k(x)}{\sum_{l=1}^K\pi_lf_l(x)}
$$
where $\pi_k$ is the prior probability that $x$ is in class number $k$, and $f_k(x)=Pr(X|Y=k)$, is the density of points $x$ in a given class $\{Y=k\}$. Assume that $f_k$ follows $\text{Norm}(\mu_k,\sigma_k)$ and that for each $k$, the variances $\sigma_k$ need not be equal. As $p=1$, each $f_k$ follows a 1-dimensional normal density given by
$$
f_k(x) = \frac{1}{\sqrt{2\pi}\sigma_k}\exp\left(-\frac{1}{2\sigma_k^2}(x-\mu_k)^2\right).
$$
Substituting into the probability equation, we have that

$$
Pr(Y=k|X=x) = \frac{\pi_k\frac{1}{\sqrt{2\pi}\sigma_k}\exp\left(-\frac{1}{2\sigma_k^2}(x-\mu_k)^2\right)}{\sum_{l=1}^K\pi_l\frac{1}{\sqrt{2\pi}\sigma_l}\exp\left(-\frac{1}{2\sigma_l^2}(x-\mu_l)^2\right)}.
$$
WLOG, we can let K=2. If show that $\sigma_1\neq\sigma_2$, leads to a non-linear decision boundary (quadratic), this suffices for the general case because for any distinct $\sigma_k\neq\sigma_l$,  the decision boundary will also be non-linear. Let $w_k=\frac{1}{2\sigma_k^2}$.

\begin{eqnarray}
Pr(Y=k|X=x) &= \frac{\pi_k\sqrt{w_k/\pi}\exp\left(-w_k(x-\mu_k)^2\right)}{\sum_{l=1}^K\pi_l\sqrt{w_k/\pi}\exp\left(-w_k(x-\mu_l)^2\right)}\\
\end{eqnarray}
the denominator of the function is the same over all classes for fixed $x$. Hence, the discriminant function $\delta_k(x)$ comes from the numerator. The decision boundary for two classes is that where $\delta_1(x)=\delta_2(x)$, hence the weight of  the denominator in this equation will cancel. We take the logarithm of the numerator to obtain
\begin{eqnarray}
\delta_k(x) &= \log\pi_k\sqrt{w_k/\pi}\exp\left(-w_k(x-\mu_k)^2\right)\\
& = \log\pi_k+\log\sqrt{w_k/\pi}-w_k(x-\mu_k)^2
\end{eqnarray}
If $\sigma_1\neq\sigma_2$, $w_1\neq w_2$ as well. Setting $\delta_1=\delta_2$, yields

\begin{eqnarray}
\delta_1(x) = \delta_2(x)\\
&\implies \log\pi_1+\log\sqrt{w_1/\pi}-w_1(x-\mu_1)^2 = \log\pi_2+\log\sqrt{w_2/\pi}-w_2(x-\mu_2)^2\\
&\implies \log(\pi_1/\pi_2)+\log\sqrt{w_1/w_2}-w_1(x-\mu_1)^2 +w_2(x-\mu_2)^2 = 0\\
&\implies \log(\pi_1/\pi_2)+\log\sqrt{w_1/w_2}+[w_1(x-\mu_1)^2 -w_2(x-\mu_2)^2] = 0\\
&\implies \log(\pi_1/\pi_2)+\log\sqrt{w_1/w_2}+[(w_1-w_2)x^2-(2w_2\mu_2-2w_1\mu_1)x+w_1\mu_1^2-w_2\mu_2^2] = 0\\
\end{eqnarray}

So we see here, that when $w_1\neq w_2$, the coefficient in front of $x^2$ will not vanish. Hence, the decision boundary for unequal variance is quadratic.

## Part a)

The curve is
$$
(1+X_1)^2+(2-X_2)^2=4.
$$
First consider the coordinate system
\begin{eqnarray}
2u = 1+X_1\\
2v = 2-X_2\\
\end{eqnarray}
To go from $(u,v)$ to $(X_1,X_2)$, we rearrange to get
\begin{eqnarray}
X_1=2u-1\\
X_2=2-2v
\end{eqnarray}

In $(u,v)$ coordinates we can rewrite the curve as 
$$
u^2+v^2=1
$$
which is the familiar circle. Let $r$ and $\theta$ be the usual polar coordinates where $r^2=u^2+v^2$ and $\theta=\arctan(v/u)$, i.e. $u=\cos(\theta)$, $v=\sin(\theta)$. This is a level curve with $r=1$ as constant. Substituting back for $X_1$ and $X_2$ we have
\begin{eqnarray}
X_1=2\cos\theta-1\\
X_2=2-2\sin\theta
\end{eqnarray}
This is a parameterization of a circle of radius 2 centered at the point $(-1,2)$. We can code this in python as follows

```python
import numpy as np
from matplotlib.pyplot import subplots, cm
# defining the parameterization taking theta into X_1 and X_2
def f(theta):
    return 2*np.cos(theta)-1,2*(1-np.sin(theta))
theta = np.linspace(0,2*np.pi,200)
f(theta)
fix, ax = subplots(figsize=(8,8))
ax.scatter(f(theta)[0],f(theta)[1]);
ax.set_title("Curve (1+X_1)^2+(2-X_2)^2=4");
```

## b)
We can add colours to the plot for points that lie inside and including the boundary of the circle, or outside of the boundary of the circle. This correspeonds to $(1+X_1)^2+(2-X_2)^2\leq4$ and $(1+X_2)^2+(2-X_2)^2 > 4$ respectively. This is essentially to say that for a point $(X_1,X_2)$, if we define a distance function $k(X_1,X_2)=(1+X_1)^2+(2-X_2)^2$, this will correspond to the regions of points such that $k>4$ and $k\leq 4$. We are finding all the points distance $\leq 4$ from the point $(-1,2)$ and labelling them one class and we are labelling the rest another class. For `python`, we can 
1. Define the function $k$
2. Generate a grid of points.
3. Label the points a colour corresponding to $k>4$ or $k\leq4$ by computing their $k$.

```python
# defining the kernel function
def k(pt):
    return ((1+pt[0])**2+(2-pt[1])**2)
x = np.linspace(-5,3,25)
y = np.linspace(-2,6,25)
box = np.meshgrid(x,y)
fig, ax = subplots(figsize=(10,8))
ax.plot(k(box),'o');
ax.set_title('Radial Distance Function');
```

```python
# function to label values of a box
def colour(x):
    # iterating through and labelling the points
    distances = k(x).flatten()
    labels = []
    for i in distances:
        if i <= 4:
            labels.append('red')
        else:
            labels.append('blue')
    return labels
# making the plot
fix, ax = subplots(figsize=(8,8))
ax.scatter(f(theta)[0],f(theta)[1],s=10,c=['red']);
ax.set_title("Curve (1+X_2)^2+(2-X_2)^2=4: In and Out");
ax.scatter(box[0],box[1],s=10,c=colour(box));
```

# Question 3: AUC and ROC
## a)

```python
import pandas as pd
# copying the data from the question
data = pd.DataFrame({'Predicted':['p','p','n','p','n','n','n','n','n','p','n'],
            'True Label': ['p','n','p','p','n','n','n','n','p','p','p']})
data
```

```python
# importing encoder to change the data into 0's as 1's
from sklearn.metrics import roc_curve, confusion_matrix
y_predict = data['Predicted']
y_true = data['True Label'] 
confuse = confusion_matrix(y_true,y_predict)
confuse
```

From the `sklearn` documentation, we know that orientation of `confusion_matrix` is 
`    Thus in binary classification, the count of true negatives is
    :math:`C_{0,0}`, false negatives is :math:`C_{1,0}`, true positives is
    :math:`C_{1,1}` and false positives is :math:`C_{0,1}``

For a confusion matrix
\begin{bmatrix}
NN&NP\\
PN&PP
\end{bmatrix}
The *true positive* rate is given as the ratio of *true positive* classifications, out of the total predictions given a "positive" label. In particular
\begin{equation}
TPR = \frac{PP}{PP+PN}
\end{equation}

```python
tpr = confuse.diagonal()[1]/sum(confuse[1,:])
```

The *false positive rate* is the rate of *false negatives* out of the total number of predictions labelled "negative", i.e.
$$
FPR = \frac{NP}{NP+NN}
$$

```python
fpr = confuse[0,1]/sum(confuse[0,:])
print("\n The TPR and FPR are {tpr:.2f}, {fpr:.2f}".format(tpr=tpr,fpr=fpr))
```

## b)
We can fill the predicted score and use the `roc_curve` function in `sklearn.metrics` to compute the AUC (Area under the Curve) of the ROC curve.

```python
data_2 = pd.DataFrame({'True Label': ['p','n','n','n','p','p','n','p','p','p'],'scores': [2.0, 6.9, 9.2, 2.8, 1.0, 7.0, 5.3, 8.1, 9.6, 1.1]})
data_2
```

```python
from sklearn.metrics import RocCurveDisplay, roc_curve, roc_auc_score
fpr, tpr, thresh = roc_curve(data_2['True Label'],data_2['scores'],pos_label='p')
figure, ax = subplots(figsize=(5,4))
ax.plot(fpr,tpr)
ax.set_xbound(-0.01,1.005)
ax.set_title('Manual ROC')
ax.set_xlabel('False positive rate')
ax.set_ylabel('True positive rate')
```

```python
# computing AUC
roc_auc_score(data_2['True Label'],data_2['scores'],labels=['p','n'])
```

```python
# plotting the roc_curve with sklearn
RocCurveDisplay.from_predictions(data_2['True Label'],data_2['scores'],pos_label='p');
```

```python
import numpy as np
from ISLP import load_data
from matplotlib.pyplot import subplots, cm
auto = load_data('Auto')
auto.columns
```

```python
mpg_med = np.median(auto['mpg'])
mpg = auto['mpg']
```

```python
# making an indicator funciton
def indicator(x,value):
    counts = np.zeros(x.shape[0])
    for i in range(len(x)):
        if (x.iloc(0)[i] > value):
            counts[i] = 1
    return counts
y = indicator(mpg,mpg_med)
y
```

We can also use the `np.where` function

```python
test = np.where(mpg>np.median(mpg),1,0)
np.equal(y,test)
```

## Part b)
We import `sklearn` to fit the linear support vector classifier for greater than median miles per gallon. We start with $C=1$, the default value and fit an initial model. Recall from part [a)](#question-4#part-a) that the variable `y` is our `mpg` indicator function. 

```python
from sklearn.svm import SVC
from ISLP.svm import plot as plot_svm
from sklearn.metrics import RocCurveDisplay
from sklearn.model_selection import train_test_split
X = auto
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=32)
```

Here we normalize our X training and test data with `StandardScaler`

```python
from sklearn.preprocessing import StandardScaler
test_1 = StandardScaler()
a = test_1.fit(X_train)
train_norm = a.fit_transform(X_train)
b = test_1.fit(X_test)
test_norm = b.fit_transform(X_test)
```

```python
svm_linear = SVC(kernel = 'linear',random_state=23)
svm_linear.fit(train_norm, y_train)
```

```python
predictions = svm_linear.predict(test_norm)
```

```python
print("Accuracy score: ",svm_linear.score(test_norm,y_test))
```

```python
roc_auc_score(y_test,predictions)
```

```python
RocCurveDisplay.from_predictions(predictions,y_test);
```
